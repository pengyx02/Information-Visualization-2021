```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

## get the original data from the apache echart
https://echarts.apache.org/examples/en/editor.html?c=watermark

```python
component={"geo": 2788,
    "title": 9575,
    "legend": 9400,
    "tooltip": 9466,
    "grid": 9266,
    "markPoint": 3419,
    "markLine": 2984,
    "timeline": 2739,
    "dataZoom": 2744,
    "visualMap": 2466,
    "toolbox": 3034,
    "polar": 1945}

charts = {"map": 3237,
    "lines": 2164,
    "bar": 7561,
    "line": 7778,
    "pie": 7355,
    "scatter": 2405,
    "candlestick": 1842,
    "radar": 2090,
    "heatmap": 1762,
    "treemap": 1593,
    "graph": 2060,
    "boxplot": 1537,
    "parallel": 1908,
    "gauge": 2107,
    "funnel": 1692,
    "sankey": 1568}
downloadJson = {'echarts.min': 17365,'echarts.simple.min': 4079,'echarts.common.min': 6929,'echarts': 14890}
const_themeJson = {'dark': 1594,'infographic': 925,'shine': 1608,'roma': 721,'macarons': 2179,'vintage': 1982}
```

## sort the data and show it in order

```python
component_sorted = sorted(component.items(),key = lambda item:item[1])#,reverse=True
charts_sorted = sorted(charts.items(),key = lambda item:item[1])
```

## detail of the optimization

+ changes of ax1 and ax2    
Make the border transparent; move the number in the x-axis; change the color
+ changes of ax3 and ax4    
<b>All numbers on the pie chart do not add up to 100 percent due to rounding</b>    
Added percentages for size comparison of different data; change the color so that different group of data have different colors;

The results are stored in a vector diagram 

```python
x = np.arange(100)
plt.rcParams['font.size'] = 14  # enlarge the size of word in the whole picture
# create four figure
fig = plt.figure(figsize = (14, 10))

ax1 = fig.add_subplot(221)
ax1.barh([i[0] for i in component_sorted], [i[1] for i in component_sorted], height=0.7, color='orange') 
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.spines['left'].set_visible(False)
plt.xticks([])
for i, v in enumerate([i[1] for i in component_sorted]):
    ax1.text(v + 3, i - .25, str(v), color='black')
   
    
ax2 = fig.add_subplot(223)
ax2.barh([i[0] for i in charts_sorted], [i[1] for i in charts_sorted], height=0.7, color='lightgreen') 
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.spines['left'].set_visible(False)
plt.xticks([])
for i, v in enumerate([i[1] for i in charts_sorted]):
    ax2.text(v + 3, i - .25, str(v), color='black')

ax3 = fig.add_subplot(222)
colors = ['lightskyblue', 'gold', 'plum', 'salmon']#tomato tomato  yellowgreen
patches,l_text,p_text = plt.pie(list(downloadJson.values()), labels=list(downloadJson.keys()), colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  # Display as a circle (avoid scaling to ellipse)
for t in p_text:
    t.set_size(14) #enlarge the size of the words, so that users can see the words clearly
    
for t in l_text:
    t.set_size(12)

ax4 = fig.add_subplot(224)
colors = ['lightskyblue', 'gold', 'plum', 'salmon','bisque','yellowgreen']#bisque salmon
patches,l_text,p_text = plt.pie(list(const_themeJson.values()), labels=list(const_themeJson.keys()), colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal') 
for t in p_text:
    t.set_size(14)

for t in l_text:
    t.set_size(12)

ax1.title.set_text('Online build')
ax3.title.set_text('Download of each version')
ax4.title.set_text('Download of each subject')

plt.savefig('echarts_plot.svg')
plt.show()
```
